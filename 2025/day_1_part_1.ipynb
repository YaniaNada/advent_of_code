```python
data = puzzle_input.copy()
print(data)
```

```python
data.rename(columns= {0: 'rotation'}, inplace= True)
```

```python
temp_list = []
for i in range(len(data)):
  temp_list.append(data['rotation'][i][0])
```

```python
data['direction'] = temp_list
```

```python
temp_list = []
for j in range(len(data)):
  temp_list.append(data['rotation'][j][1:])
```

```python
data['steps']= temp_list
data['steps'] = data['steps'].astype(int)
```

```python
data
```

```python
initial_value = 50
current_value = initial_value
new_value = []
for i in range(len(data)):

  step = data['steps'][i]

  if data['direction'][i] == 'L':
    for j in range(step):
      if current_value <= 0:
        current_value = 100
        current_value -=1
      else:
        current_value -=1

  elif data['direction'][i] == 'R':
      for j in range(step):
        current_value +=1
        if current_value >= 99:
          current_value = -2
          current_value +=1

  print(f'current value: {current_value}, {i},{j}, {data['direction'][i]}')
  new_value.append(current_value)


```

```python
count=0
for i in range(len(new_value)):
  if new_value[i] == 0:
    count += 1
    print(f'{i}, {new_value[i]}')
```
