```python
import pandas as pd
```

```python
data = pd.read_excel('advent_of_code_2024.xlsx', sheet_name=1, header=None)
data
```

```python
sample= data.copy()
```

```python
for i in range(len(sample)):

    x = sample.iloc[i]
    y = len(sample.iloc[i])-2
    j=0

    for j in range(y):

        if (abs(x[j] - x[j+1]) > 3)or(abs(x[j] - x[j+1]) == 0):
            sample.loc[i,'pattern'] = 'unsafe'
            break
        
        elif (abs(x[j] - x[j+1]) <= 3):
            
            is_increasing = True
            is_decreasing = True
            
            for n in range(y):
                if x[n] < x[n + 1]:
                    is_decreasing = False
                elif x[n] > x[n + 1]:
                    is_increasing = False
            
            if is_increasing or is_decreasing:
                sample.loc[i,'pattern'] = 'safe'
                
            else:
                sample.loc[i,'pattern'] = 'unsafe'

```
